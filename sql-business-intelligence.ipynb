````python
# Create database
conn = sqlite3.connect('retail_business.db')
cursor = conn.cursor()

# Create Customers table
cursor.execute('''
CREATE TABLE IF NOT EXISTS customers (
    customer_id INTEGER PRIMARY KEY,
    customer_name TEXT,
    email TEXT,
    city TEXT,
    state TEXT,
    region TEXT,
    segment TEXT,
    join_date TEXT
)''')

# Create Products table
cursor.execute('''
CREATE TABLE IF NOT EXISTS products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    sub_category TEXT,
    unit_price REAL
)''')

# Create Orders table
cursor.execute('''
CREATE TABLE IF NOT EXISTS orders (
    order_id INTEGER PRIMARY KEY,
    customer_id INTEGER,
    product_id INTEGER,
    order_date TEXT,
    quantity INTEGER,
    discount REAL,
    sales REAL,
    profit REAL,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id)
)''')

conn.commit()
print("✅ Database and tables created successfully!")
````

````python
query1 = '''
SELECT 
    COUNT(DISTINCT o.order_id) AS total_orders,
    COUNT(DISTINCT o.customer_id) AS unique_customers,
    ROUND(SUM(o.sales), 2) AS total_revenue,
    ROUND(SUM(o.profit), 2) AS total_profit,
    ROUND((SUM(o.profit) / SUM(o.sales)) * 100, 2) AS profit_margin_pct
FROM orders o
'''

df_q1 = pd.read_sql_query(query1, conn)
print("=== BUSINESS OVERVIEW ===")
print(df_q1.to_string(index=False))
````

````python
query2 = '''
SELECT 
    p.category,
    COUNT(o.order_id) AS total_orders,
    ROUND(SUM(o.sales), 2) AS total_revenue,
    ROUND(SUM(o.profit), 2) AS total_profit,
    ROUND((SUM(o.profit) / SUM(o.sales)) * 100, 2) AS profit_margin_pct
FROM orders o
JOIN products p ON o.product_id = p.product_id
GROUP BY p.category
ORDER BY total_revenue DESC
'''

df_q2 = pd.read_sql_query(query2, conn)
print("=== REVENUE BY CATEGORY ===")
print(df_q2.to_string(index=False))
````

````python
query3 = '''
SELECT 
    p.product_name,
    p.category,
    COUNT(o.order_id) AS times_ordered,
    SUM(o.quantity) AS total_units_sold,
    ROUND(SUM(o.sales), 2) AS total_revenue,
    ROUND(SUM(o.profit), 2) AS total_profit
FROM orders o
JOIN products p ON o.product_id = p.product_id
GROUP BY p.product_name, p.category
ORDER BY total_revenue DESC
LIMIT 5
'''

df_q3 = pd.read_sql_query(query3, conn)
print("=== TOP 5 BEST SELLING PRODUCTS ===")
print(df_q3.to_string(index=False))
````

````python
query4 = '''
SELECT 
    c.region,
    COUNT(DISTINCT o.customer_id) AS unique_customers,
    COUNT(o.order_id) AS total_orders,
    ROUND(SUM(o.sales), 2) AS total_revenue,
    ROUND(SUM(o.profit), 2) AS total_profit,
    ROUND(SUM(o.sales) / COUNT(o.order_id), 2) AS avg_order_value
FROM orders o
JOIN customers c ON o.customer_id = c.customer_id
GROUP BY c.region
ORDER BY total_revenue DESC
'''

df_q4 = pd.read_sql_query(query4, conn)
print("=== REVENUE BY REGION ===")
print(df_q4.to_string(index=False))
````

````python
query5 = '''
SELECT 
    strftime('%Y', order_date) AS year,
    strftime('%m', order_date) AS month,
    COUNT(order_id) AS total_orders,
    ROUND(SUM(sales), 2) AS monthly_revenue,
    ROUND(SUM(profit), 2) AS monthly_profit
FROM orders
GROUP BY year, month
ORDER BY year, month
'''

df_q5 = pd.read_sql_query(query5, conn)
print("=== MONTHLY REVENUE TREND ===")
print(df_q5.to_string(index=False))
````

````python
query6 = '''
SELECT 
    c.segment,
    COUNT(DISTINCT o.customer_id) AS unique_customers,
    COUNT(o.order_id) AS total_orders,
    ROUND(SUM(o.sales), 2) AS total_revenue,
    ROUND(SUM(o.sales) / COUNT(DISTINCT o.customer_id), 2) AS revenue_per_customer
FROM orders o
JOIN customers c ON o.customer_id = c.customer_id
GROUP BY c.segment
ORDER BY total_revenue DESC
'''

df_q6 = pd.read_sql_query(query6, conn)
print("=== CUSTOMER SEGMENT ANALYSIS ===")
print(df_q6.to_string(index=False))
````

````python
query7 = '''
SELECT 
    CASE 
        WHEN discount = 0 THEN 'No Discount'
        WHEN discount <= 0.1 THEN '10% Discount'
        WHEN discount <= 0.2 THEN '20% Discount'
        ELSE '30%+ Discount'
    END AS discount_tier,
    COUNT(order_id) AS total_orders,
    ROUND(SUM(sales), 2) AS total_revenue,
    ROUND(SUM(profit), 2) AS total_profit,
    ROUND((SUM(profit) / SUM(sales)) * 100, 2) AS profit_margin_pct
FROM orders
GROUP BY discount_tier
ORDER BY profit_margin_pct DESC
'''

df_q7 = pd.read_sql_query(query7, conn)
print("=== DISCOUNT IMPACT ON PROFIT ===")
print(df_q7.to_string(index=False))
````

````python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('SQL Business Intelligence Dashboard', 
             fontsize=16, fontweight='bold')

colors = ['#00d4aa', '#4f8fff', '#ffd166', '#ff6b6b']

# Chart 1 - Revenue by Category
axes[0,0].bar(df_q2['category'], df_q2['total_revenue'], 
              color=colors, edgecolor='white')
axes[0,0].set_title('Revenue by Category', fontweight='bold')
axes[0,0].set_ylabel('Revenue ($)')
for p in axes[0,0].patches:
    axes[0,0].annotate(f'${p.get_height():,.0f}',
                       (p.get_x()+p.get_width()/2, p.get_height()+1000),
                       ha='center', fontsize=9)

# Chart 2 - Revenue by Region
axes[0,1].bar(df_q4['region'], df_q4['total_revenue'],
              color=colors, edgecolor='white')
axes[0,1].set_title('Revenue by Region', fontweight='bold')
axes[0,1].set_ylabel('Revenue ($)')

# Chart 3 - Revenue by Segment
axes[1,0].pie(df_q6['total_revenue'], labels=df_q6['segment'],
              colors=colors[:3], autopct='%1.1f%%',
              startangle=90)
axes[1,0].set_title('Revenue by Customer Segment', fontweight='bold')

# Chart 4 - Discount Impact
axes[1,1].bar(df_q7['discount_tier'], df_q7['profit_margin_pct'],
              color=colors, edgecolor='white')
axes[1,1].set_title('Profit Margin by Discount Tier', fontweight='bold')
axes[1,1].set_ylabel('Profit Margin (%)')
axes[1,1].set_xticklabels(df_q7['discount_tier'], rotation=15)

plt.tight_layout()
plt.savefig('sql_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved as sql_dashboard.png")
````
